## Product Recommender

```python
df['tags'] = df[['name', 'sub_category']].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

##Text Vectorization
###Technique using --> (Bag of words) Text similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

##Finding cosine distance

```python
# checking similarity b/w vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculate distance/similarity b/w vectors
similarity = cosine_similarity(vectors)
```

```python
from fuzzywuzzy import fuzz

def recommend(product):
    # Perform fuzzy matching to find similar product names
    ratios = df['name'].apply(lambda x: fuzz.partial_ratio(x.lower(), product.lower()))
    matched_indices = ratios.sort_values(ascending=False).index[:12]

    data = []
    for index in matched_indices:
        item = []
        product_name = df.loc[index, "name"]
        product_image = df.loc[index, "image"]
        product_price = df.loc[index, "actual_price"]
        product_ratings = df.loc[index, "ratings"]
        product_total_ratings = df.loc[index, "no_of_ratings"]
        product_buy_link = df.loc[index, "link"]

        item.extend(["Prod.Name:", product_name])
        item.extend(["Prod.Image:", product_image])
        item.extend(["Prod.Price:", product_price])
        item.extend(["Prod.Ratings:", product_ratings])
        item.extend(["Prod.Total Ratings:", product_total_ratings])
        item.extend(["Prod.Buy:", product_buy_link])

        data.append(item)

    return data
```

```python
recommend('samsung Smart TV')
```
